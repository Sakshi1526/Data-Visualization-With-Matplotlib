# Box Plot

1) It is used to represent the quartile/ percentile distribution of data
2) It is used to visualize distribution of numerial data
3) It represents a5 point summary
   
   Q1,Q2,Q3,min and max

i) <b>Q1</b> : 25 percentile</br>
ii) <b>Q2</b> : 50 percentile(median)</br>
iii) <b>Q3</b> : 70 percentile</br>
iv) <b>uqr (Inter Quartile Range</b> : Q3 - Q1</br>
v) <b>min</b> : q1 - 1.5 IQR</br>
vi) <b>max</b> : q3 - 1.5 IQR</br>
vii) <b>Outliers</b> : >max or <min

### Demo Example

```python
x = [12,14,16,15,14,25,12,11,9,10,13,15,2,17,9,16]
print(x)
print(len(x))
```

```python
plt.boxplot(x)
plt.grid()
plt.show()

# q3 = 15.25
# q2 = 13.5
# q1 = 10.75
# max = 22 (actual)
# min = 4 (actual)
# max in the graph  = value closest to 22 less than outlier 25 in the list x = 17
# min in the graph  = value closest to 2 greater than outlier 2 in the list x = 9
```

```python
q1 = np.percentile(x,25)
q2 = np.percentile(x,50)   # Median
q3 = np.percentile(x,75)
iqr = q3-q1              # Inter Quartile Range
print('Q1',q1)
print('Q2',q2)  # Median
print('Q3',q3)
print('IQR',iqr)
bmax = q3 + 1.5 * iqr
bmin = q1 - 1.5 * iqr
print('Max',bmax)
print('Min',bmin)
x.sort()
print(x)
```

# Stack Bar Chart

### Q3) Plot mean of Budget and Profit based on a genre1 on a stacked area chart

```python
r1 =  d1.groupby('genre_1')[['budget','Profit']].mean()
r1.index.names=['Genre']
r1.head()
```

#### Overlapped values

```python
plt.bar(r1.index,r1['budget'],color='cyan',label='Mean Budget')
plt.bar(r1.index,r1['Profit'],color='orange',label='Mean Profit')
plt.xlabel('Genre')
plt.legend()
plt.title('Genre Wise Mean Budget and Profit')
plt.xticks(rotation = 90)
plt.grid()
plt.ylim(0,280)
plt.show()
```

#### Using the bottom parameter to create stacked bar charts

```python
plt.bar(r1.index,r1['budget'],color='cyan',label='Mean Budget')
plt.bar(r1.index,r1['Profit'],color='orange',label='Mean Profit',bottom = r1['Profit']) 
plt.xlabel('Genre')
plt.legend()
plt.title('Genre Wise Mean Budget and Profit')
plt.xticks(rotation = 90)
plt.grid()
plt.ylim(0,280)
plt.show()
```

```python
r1.plot(kind ='bar', stacked = True)
plt.grid()
plt.show()
```

# Stacked Area Chart

Multiple column data values when represented into area chart are placed on top of another area chart is referred to as  stacked area chart

<b>plt.stackplot(y1,y2,y3,.....yn,labels)</b><br>

x - data (column) usually categorical</br>
y1,y2,y3,....yn - columns values that needs to be stacked.<br>
labels -  corresponding labels to y1,y2,y3,....yn. It is a list. </br>

```python
plt.stackplot(r1.index, r1 ['budget'], r1['Profit'],labels  = ['Budget','Profit'])
plt.xlabel('Genre1')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
r2 = d1.groupby(['genre_1'])[['budget','Gross','Profit']].mean()
r2.index.names = ["Genre"]
r2
```

```python
plt.stackplot(r2.index, r2 ['budget'],r2['Gross'], r2['Profit'],
              labels  = ['Budget','Gross','Profit'])
plt.xlabel('Genre1')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
