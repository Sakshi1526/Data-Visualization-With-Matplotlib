# Bar Chart

1) The graph represents categories on 1 axis and a continuous value in the other.
2) It can be used to compare different categories of data.

#### Syntax

###### plt.bar(x,y,color,label,edgecolor)
x - data(column) to be represeted on the x-axis</br>
y - data(column) to be represeted on the y-axis</br>
color - used ise sed to specify color of the bar</br>
edgecolor -used to define to color of the edge of the bar chart

# Q1) Plot year wise average Runtime of top 7 movies on a bar chart

```python
q1 = d1.groupby('title_year')['Runtime'].mean().reset_index()
q1 = q1.sort_values(by = 'title_year', ascending = True).head(7)
q1
```

```python
plt.bar(q1['title_year'],q1['Runtime'])
```

# Graph Customization

## 1) Add xlabel, ylabel and title, show

### a) plt.xlabel()
It is used to provide to the x-axis

### b) plt.ylabel()
It is used to provide to the y-axis

### c) plt.title()
It is used to provide title to the chart

### d) plt.show()
This will ensure that all of the figures are shownn and this function immediately returns. It gets rid of the gibberish data displayed above the figure.



```python
plt.bar(q1['title_year'],q1['Runtime'])
plt.title('Year wise avg Runtime', fontsize = 15, color = 'Red')
plt.xlabel('Tile_Year')
plt.ylabel('Runtim')
plt.show()
```

## 2) Add color and edge color to the bars

### e) color
It s an attribute used to define the color of the bars in chart.</br>
A single color can be chosen or a list of colors can be passed as an argument.

### f) edgecolor
It is an attribute used to define the edgecolor of the bars in bar chart

```python
plt.figure(figsize = (8,4))
plt.bar(q1['title_year'],q1['Runtime'], color = 'orange', edgecolor = 'black')
plt.xlabel('Tile_Year')
plt.ylabel('Runtime')
plt.title('Year wise mean Runtime', fontsize = 16, color = 'Blue')
plt.show()
```

```python
plt.bar(q1['title_year'],q1['Runtime'], color = ['orange','lightblue','violet','lightgreen','brown','tan','yellow'],
        edgecolor = 'black')
plt.xlabel('Tile_Year')
plt.ylabel('Runtime')
plt.title('Year wise avg Runtime', fontsize = 16, color = 'Blue')
plt.show()
```

# Q2) Plot genre_1 wise mean of budget and gross on bar chart

Converting Budget and gross from dollars to milliom dollars

```python
d1["Gross"] = d1["Gross"]/1000000
d1["budget"] = d1["budget"]/1000000
d1.head()
```

```python
d1.columns
```

```python
q2 = d1.groupby('genre_1')[['budget','Gross']].mean()
print(type(q2))
q2.index.names = ['Genre']
q2.head()
```

## 3) Add label and legend

### g) label
It is a parameter of plt.bar() function . It is used to assign a distinct label to each of the categories.

### h) plt.legend()
Legend ensures that the label parameter added gets reflected in the chart

### alpha
It is parameter of bar chart used to set the transparancy

```python
print(q2.index)
# x axis having index
```

```python
plt.bar(q2.index,q2['budget'],label = 'Budget', color ='blue')
plt.bar(q2.index,q2['Gross'],label = 'Gross', color ='maroon', alpha = 0.5)
plt.xlabel('Genre')
plt.title('Genre Wise Mean Budget and Gross')
plt.legend(loc = 0)
plt.show()
```

## 4) Add xticks, yticks, aplha

### i) plt.ticks()
1) It can be used to rotate the label in case they are overlapping on the x-axis values
2) It can also be used to assign some alias names to the labels on x-axis values, that is it can be used to change the axis scale and ticks

### j) plt.ticks()
1) It can be used to rotate the label in case they are overlapping on the y-axis values
2) It can also be used to assign some alias some names to the labels on y-axis

### k) aplha
It is used to set transparenccy.Used to distinguish the bars using transparency when they overlap

```python
print(q2.index)
```

```python
plt.bar(q2.index,q2['budget'],label = 'Budget', color ='red')
plt.bar(q2.index,q2['Gross'],label = 'Gross', color ='green', alpha = 0.5)
plt.xlabel('Genre')
plt.title('Genre Wise Mean Budget and Gross')
plt.legend()
plt.xticks(rotation = 30)
plt.yticks(rotation =45)
plt.show()
```

### Changing the axis ticks and axis scale in the above graph

```python
plt.bar(q2.index,q2['budget'],label = 'Budget', color ='red')
plt.bar(q2.index,q2['Gross'],label = 'Gross', color ='green', alpha = 0.5)
plt.xlabel('Genre')
plt.title('Genre Wise Mean Budget and Gross')
plt.legend()
plt.xticks(q2.index,['G1','G2','G3','G4','G5','G6','G7','G8'])
plt.show()
```

# Q3) Plot actor_1_name based and their mean of fb_likes for the top 7 actors with highest fb_likes on a bar chart

```python
q3 = d1.groupby('actor_1_name')['actor_1_facebook_likes'].mean().reset_index()
q3 = q3.sort_values(by = 'actor_1_facebook_likes',ascending  = False).head(7)
q3.columns = ['Actor','Fb_Likes']
q3
```

```python
plt.bar(q3['Actor'],q3['Fb_Likes'],color ='pink', label = 'Fb_Likes', edgecolor = 'black')
plt.xlabel('Actor')
plt.ylabel('Fb_Likes')
plt.title('Actor Wise Fb_Likes')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

## 5) Adding xlim, ylim, figure, savefig

### l) plt.xlim()
It is used to provide range of values for the x-axis

### m) plt.ylim()
It is used to provide range of values for the y-axis

### n) plt.figure(figsize)
It is used to change the figure size

### o) plt.savefig
It is used to save the figure

```python
plt.figure(figsize=(8,5))
plt.bar(q3['Actor'],q3['Fb_Likes'],color ='tan', label = 'Fb_Likes', edgecolor = 'black')
plt.xlabel('Actor')
plt.ylabel('Fb_Likes')
plt.title('Actor Wise Fb_Likes')
plt.xticks(rotation = 90)
plt.legend()
plt.ylim(0,45000)
plt.savefig('Actor_vs_Fb_likes14Aug.png')
plt.show()
```

## Adding text, annotation

### p) plt.text(x,y,s,color,fontsize)
It is used to add text in the chart

### q) plt.annotate()
It is used to add annotation to the chart

#### Adding Text

```python
plt.figure(figsize=(8,5))
plt.bar(q3['Actor'],q3['Fb_Likes'],color ='Lightblue', label = 'Fb_Likes', edgecolor = 'black')
plt.text(2,15000,'Actor FB Likes', color ='Blue', fontsize = 18)
plt.xlabel('Actor')
plt.ylabel('Fb_Likes')
plt.title('Actor Wise Fb_Likes')
plt.xticks(rotation = 60)
plt.legend()
plt.show()
```

#### Adding Annotations

```python
s = list(range(7))    # s =[0,1,2,3,4,5,6]
plt.bar(s,q3['Fb_Likes'],color='orange', label= 'Fb_likes', edgecolor = 'black')
plt.ylim(0,45000)
for idx, val in enumerate(q3['Fb_Likes']):     # gernate additional counter (idx)
    plt.annotate(xy = [s[idx] -0.4, int(val) + 1000], s = int(val))
plt.xticks(s,q3['Actor'],rotation =90)
plt.legend()
plt.show()
```

# Q4) Plot mean Budget  vs mean Gross on Bar chart for different years. Represent Budget and Gross on 2 different axis

```python
q4 = d1.groupby('title_year')[['budget','Gross']].mean()
q4
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)
ax2 = ax.twinx()    # Create another axes that shares the same x-axis as ax.

width = 0.4

q4['budget'].plot(kind ='bar',color ='purple', ax=ax, width = width, position=1)
q4['Gross'].plot(kind ='bar',color ='orange', ax=ax2, width = width, position=0)

ax.set_ylabel('Budget')
ax2.set_ylabel('Profit')
plt.show()
```
