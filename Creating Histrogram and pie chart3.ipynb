# Histrogram
1) It is used to represent frequency distribution
2) One axis represents the variables in form of bars, the other the freqency of each bar

#### plt.hist(x,color,bins)

x -  data(column) for which we need to compute the frequency<br>
color -Used to specify the color of the bar of histrogram<br>
bins - Buckets of which the frequency will be computed<br>
edgecolor -Used to specify the edgecolor of the bars of histrogram


```python
x = [1,2,3,2,4,2,3,1,5,6,4,4,3,5,7,3,6,7,2,3,4,7]
print(len(x))
```

```python
from collections import Counter
c1 = Counter(x)
print(c1)
```

```python
plt.hist(x,color='orange')
plt.show()
```

```python
plt.hist(x,bins=[1,2,3,4,5,6,7],color='pink',edgecolor='black')
print(c1)
plt.show

# 1st bin : [1,2) => Including count pf 1, excluding count of 2 => 2
# 2nd bin : [2,3) => Including count pf 2, excluding count of 3 => 4
# 3rd bin : [3,4) => Including count pf 3, excluding count of 4 => 5
# 4th bin : [4,5) => Including count pf 4, excluding count of 5 => 4
# 5th bin : [5,6) => Including count pf 5, excluding count of 6 => 2
# 6th bin : [6,7) => Including count pf 6, excluding count of 7 => 2
# 7th bin : [7,8] => Including count pf 7, excluding count of 8 => 3
```

# Pie Chart
1) It is used to represent precentage distribution.
2) To plot pie chart, we need a continous column and its corresponding labels (the categories that this continous column highlights)

###  plt.pie(x,labels,color,explode,autopct,shadow)

x - data(continuous column) for which percentage values will be computed<br>
labels - Categorical description for the continuous column x. It will be represented as legend<br>
color - Used to specify the color of the different pies<br>
autopct - Used to display the precentage values in a particular format<br>
shadow - Used to set the shadow on or off<br>
counterclock - Used to display the values anitclockwise or clockwise<br>

## Q4) Represent the percentage distribution of mean of profit based on title_year year on a pie chart

Where Profit = Gross - Budget

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head(3)
```

```python
q3 = d1.groupby('title_year')['Profit'].mean().reset_index()
q3
```

```python
sum_profit = q3['Profit'].sum()
print(sum_profit)


per_2015 = 107.508278/sum_profit
print(per_2015*100)
```

```python
plt.pie(x = q3['Profit'],labels=q3['title_year'],autopct ='%.2f%%')
plt.show()
```

```python
plt.pie(x = q3['Profit'],labels=q3['title_year'],
       colors = ['red','blue','yellow','tan','brown','green','cyan'],autopct = '%.2f%%',
       explode=(0,0,0,0,0,0,0.1))
plt.title('Year wise mean of profit in million dollars', fontsize = 15, color='blue')
plt.show()
```

```python
plt.pie(x = q3['Profit'],labels=q3['title_year'],
       colors = ['red','blue','yellow','tan','brown','green','cyan'],autopct = '%.2f%%',
       explode=(0.1,0,0.2,0,0,0,0))
plt.title('Year wise mean of profit in million dollars', fontsize = 15, color='blue')
plt.show()
```

```python
plt.pie(x = q3['Profit'],labels=q3['title_year'],
       colors = ['red','blue','yellow','tan','brown','green','cyan'],autopct = '%.2f%%',
       explode=(0,0,0,0,0,0,0.1),shadow = True)
plt.title('Year wise mean of profit in million dollars', fontsize = 15, color='blue')
plt.show()
```

```python
plt.pie(x = q3['Profit'],labels=q3['title_year'],
       colors = ['red','blue','yellow','tan','brown','green','cyan'],autopct = '%.2f%%',
       explode=(0,0,0,0,0,0,0.1),shadow = True, counterclock =False)
plt.title('Year wise mean of profit in million dollars', fontsize = 15, color='blue')
plt.show()
```
