### Converting budget and Gross into Million dollars

```python
d1["Gross"] = d1["Gross"]/1000000
d1["budget"] = d1["budget"]/1000000
d1.head()
```

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head(3)
```

# Multiple Plots

Multiple plots can be plotted using Subplots</b>

<b>fig,ax = plt.subplots(nrows,ncols,figsize,shareX,shareY)<b></br>

It is used to multiple plots in a single chart area.</br>

nrows - Defines the number of rows</br>
ncols - Defines the number of columns</br>
figsize - used to define the figure size</br>
shareX - Set to True, when common X-axis is to be set amongst multiple charts</br>
shareY - Set to true, when common Y-axis is to be set amongst multiple charts</br>


### Q1) Plot sum of Imdb_rating and mean of MetaCritic based on year using subplots

```python
r1 = d1.groupby('title_year').agg({'IMDb_rating':['sum'],'MetaCritic':['mean']})
r1
```

### Multiple plots without subplot

```python
print(r1.index)
print(r1.columns)
```

### Q2) Plotting IMDB_rating sum of bar chart and on a line chart in the same plot (without Subplots)

```python
plt.bar(r1.index,r1[('IMDb_rating','sum')],color = 'orange', edgecolor = 'black')
plt.plot(r1.index,r1[('IMDb_rating','sum')],color = 'maroon', lw =3, marker='o')
plt.xlabel('Year')
plt.ylabel('Sum IMDB_Rating')
plt.title('Sum IMDB_Rating vs Year')
plt.show()
```

```python
print(r1.columns)
```

### Q3) Plot sum of Imdb_rating and mean of MetaCritic based on year using subplots

a) Sum of Imdb_rating vs year on a bar chart.</br>
b) Mean od MetaCritic vs year on a line chart

```python
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(5,8))    # nrows = 2, ncols = 1
# ax1 is the name of plot1
# ax2 is the name of plot2

ax1.bar(r1.index,r1[('IMDb_rating','sum')],color = 'lightgreen', edgecolor = 'black')
ax1.set_title('Sum IMDB_Rating vs Year')
ax1.set_ylabel('Sum of IMDB_Rating')
ax1.set_xlabel('Year')

ax2.plot(r1.index,r1[('MetaCritic','mean')],color = 'maroon', marker = 'd', markersize = 8)
ax2.set_title('Mean MetaCritic vs Year')
ax2.set_ylabel('Mean Meta Critic')
ax2.set_xlabel('Year')

plt.show()
```

### Q4) Create a dataframe cinsisting of Title, budget,profit, Imdb_rating, Runtime, content_rating, actor1_name, genre1. Filter the resukt based on top 10 profitable movies.

```python
r2 = d1[['Title','budget','Profit','IMDb_rating', 'Runtime','content_rating','actor_1_name','genre_1']]
r2 = r2.sort_values(by = 'Profit', ascending = False).head(10)
r2
```

```python
r2['Title'] = r2['Title'].str.slice(0,22)
r2.head()
```

### Q4) From r2 dataframe, generate content rating wise mean of IMDB_rating

```python
r3 = r2.groupby(['content_rating'])['IMDb_rating'].mean()
r3.head()
```

### (2,2) Subplots based on following charts

a) Title vs Profit on a horizontal bar chart from r2 dataframe</br>
b) Pie chart of Mean IMDB rating based on content from r3 dataframe</br>
c) Boxplot of Budget from r2 dataframe</br>
d) Histrogram of Runtime from r2 dataframe

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(16,12) )  # nrows = 2, ncols = 2
# ax1 is row1,col1, ax2 is row1, col2
# ax3 is row2,col1, ax4 is row1, col2

ax1.bar(r2['Title'],r2['Profit'],color = 'pink', edgecolor = 'black')
ax1.set_title('Title vs Profit')
ax1.grid()

ax2.pie(r3.values, labels = r3.index, autopct = '%.2f%%')
ax2.set_title('Percentage distribution of Content rating')

ax3.boxplot(r2['budget'])
ax3.set_title('Quartile distribution of Budget')

ax4.hist(r2['Runtime'], bins =list(range(r2['Runtime'].min(), r2['Runtime'].max())))
ax4.set_title('Frequency distribution of Runtime')

plt.show()
```
