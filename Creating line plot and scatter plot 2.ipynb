# Line Chart
1) It is used to represent the variation of a continous data over time
2) Howerver, it can be used with categorical columns as well

#### plt.plot(x,y,color,lw,label,linestyle,marker)

x - data(column) to be plotted on the x-axis<br>
y - data(column) to be plotted on the y-axis<br>
color - color of the line<br>
lw - linewdith<br>
linestyle - types of line  <br>
marker - highlights the connection between x and y<br>
markersize - Determines the szie of the marker

Types of Markers: * , o, 1, 2, 3, d, <, >, ^ etc<br>
Types of Linestyle:    --(dashed), -.(dash dotted), :(dotted)

# Q1) Plot count of genre2 on a line chart where count>1

```python
q1 = df['genre_2'].value_counts()
q1 = q1[q1>1]
q1
```

```python
print(q1.index)
print(q1.values)
```

```python
plt.figure(figsize =(9,4))
plt.plot(q1.index,q1.values,color='maroon',marker='o',
         label='count',markersize=5,lw=2,linestyle='--')
plt.xlabel('Genre2')
plt.ylabel('Count')
plt.legend()
plt.title('Genre2 vs Count', fontsize =15, color ='blue')
plt.grid()
plt.show()
```

# Q2) Plot year wise mean of Profit on a line chart where Profit = Gross - budget

Converting Budget and Gross from dollars to Millon dollars

```python
d1["Gross"] = d1["Gross"]/1000000
d1["budget"] = d1["budget"]/1000000
d1.head()
```

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head()
```

```python
q2 = d1.groupby('title_year')['Profit'].mean().reset_index()
q2 = q2.sort_values(by='title_year',ascending=True)
q2
```

```python
plt.plot(q2['title_year'],q2['Profit'],color ='brown',marker='*')
plt.title('Year wise avg Profit(in M)', fontsize =10, color='blue')
plt.xlabel('Title_Year')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
q3 = d1.groupby('title_year')[['budget','Profit']].mean().reset_index()
q3 = q3.sort_values(by='title_year',ascending=True)
q3
```

```python
plt.plot(q3['title_year'],q3['budget'],color ='green',marker='<',label='Budget',markersize = 6)
plt.plot(q3['title_year'],q3['Profit'],color ='red',marker='>',label='Profit',markersize=5)
plt.title('Year wise Mean Budget And Profit(in M)', fontsize =10, color='blue')
plt.xlabel('Title_Year')
# plt.ylabel('Profit')
plt.legend()
plt.grid()
plt.show()
```

# Scatter Chart
1) It is used to represent the distribution of the data<br>
2) It gives us the idea of covariance amongst the continuous fields plotted<br>
3) It is only meant for continuous data(columns)<br>

#### plt.scatter(x,y,color,label,marker,s)
    
x - data(column) to be plotted on the x-axis<br>
y - data(column) to be plotted on the y-axis<br>
color - color of the line<br>
marker - highlights the connection between x and y<br>
s - Determines the szie of the marker

Types of Markers: * , o, 1, 2, 3, d, <, >, ^ etc<br>

# Q3) Plot a scatter chart between Profit and Budget

```python
plt.scatter(d1['Profit'],d1['budget'],marker='d',color='Purple',s=14)
plt.xlabel('Profit')
plt.ylabel('Budget')
plt.title('Profit Vs Budget', fontsize = 15, color ='red')
plt.grid()
plt.show()
```

```python
plt.scatter(d1['Profit'],d1['budget'],marker='^',s=50,color='pink',label='Profit Vs Budget')
plt.scatter(d1['Profit'],d1['Gross'],marker='o',s=45,color='lightblue',label='Profit Vs Gross')
plt.xlabel('Profit')
plt.title('Profit Vs Budget and Gross', fontsize = 15, color ='Black')
plt.grid()
plt.legend()
plt.show()
```
